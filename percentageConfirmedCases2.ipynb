```python
import pandas as pd
import numpy as np
import datetime
from matplotlib import pyplot as plt
from pylab import * 
import matplotlib.ticker as mticker
#time-series-19-covid-selectcountries
#brazil, india, 
#comparing Egypt, Indonesia, Iran, Mexico, Peru, UK


dateparse = lambda x:pd.datetime.strptime(x,'%m/%d/%Y')
egyptcases = pd.read_csv('egypt.csv', parse_dates=['Date'], date_parser=dateparse)
indonesiacases = pd.read_csv('indonesia.csv', parse_dates=['Date'], date_parser=dateparse)
irancases = pd.read_csv('iran.csv', parse_dates=['Date'], date_parser=dateparse)
mexicocases = pd.read_csv('mexico.csv', parse_dates=['Date'], date_parser=dateparse)
perucases = pd.read_csv('peru.csv', parse_dates=['Date'], date_parser=dateparse)
ukcases = pd.read_csv('uk.csv', parse_dates=['Date'], date_parser=dateparse)

######################################################################################

fig= plt.figure(figsize=(12,8))

egx = egyptcases.Date
egy = egyptcases.Confirmed  
inx = indonesiacases.Date
iny = indonesiacases.Confirmed  
irx = irancases.Date
iry = irancases.Confirmed 
mxx = mexicocases.Date
mxy = mexicocases.Confirmed 
prx = perucases.Date
pry = perucases.Confirmed 
ukx = ukcases.Date
uky = ukcases.Confirmed 

egyptcases = egyptcases.set_index('Date')
indonesiacases = indonesiacases.set_index('Date')
irancases = irancases.set_index('Date')
mexicocases = mexicocases.set_index('Date')
perucases = perucases.set_index('Date')
ukcases = ukcases.set_index('Date')

plt.rcParams["font.family"] = "Times New Roman"
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')

#comparing Egypt, Indonesia, Iran, Mexico, Peru, UK
plt.fill_between(egx,egy,  alpha=0.1, color = '#B64C07')
plt.fill_between(inx,iny, alpha=0.1, color = '#063957')
plt.fill_between(irx,iry, alpha=0.1, color = '#BB5C0D')
plt.fill_between(mxx,mxy,  alpha=0.1, color = '#006400')
plt.fill_between(prx,pry,  alpha=0.1, color = '#FFC300')
plt.fill_between(ukx,uky,  alpha=0.1, color = '#060649')

ax = gca()

ax.xaxis.tick_bottom()
ax.yaxis.tick_left()


ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')


x_ticks = np.arange(0,3000000,500000)

#comparing Egypt, Indonesia, Iran, Mexico, Peru, UK
plt.plot(egx, egy, color = '#B64C07', label = 'Egypt', linewidth=1.8)
plt.plot(inx, iny, color = '#063957', label = 'Indonesia', linewidth=1.8)
plt.plot(irx, iry, color = '#f48670', label = 'Iran', linewidth=1.8)
plt.plot(mxx, mxy, color = '#006400', label = 'Mexico', linewidth=1.8)
plt.plot(prx, pry, color = '#FFC300', label = 'Peru', linewidth=1.8)
plt.plot(ukx, uky, color = '#8A5AA0', label = 'UK', linewidth=1.8)


plt.legend()
plt.show()
```
